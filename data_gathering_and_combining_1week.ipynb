How do the features in the first 8 weeks relate to the defensive formation? This could give us a baseline for the other half of the season. Because of this, we will add two new features: 'defender' and 'coverage_count'. 

```python
#Gather the first week if play into one data frame 
data_path = '../nfl_data/'
player_tracking=[]
#range(1,2) will give us the first week of player tracking data 
for k in range(1,2):
    player_tracking.append(pd.read_csv(data_path+'week{}.csv'.format(k)))
player_tracking_df = pd.concat(player_tracking)
```

```python
player_tracking_df
```

```python
#create a new unique Id to help get the eventual defender count
player_tracking_df['new_Id'] = player_tracking_df['gameId'].astype(str) + '_' + player_tracking_df['playId'].astype(str)
player_tracking_df.sample(10)
```

```python
player_tracking_df.position.unique()
```

```python
rel_players=['QB', 'SS', 'WR', 'FS', 'RB', 'MLB', 'CB', 'TE', 'LB', 'FB','OLB', 
             'HB', 'ILB', 'DL', 'DB', 'NT', 'S', 'DE','DT']
```

```python
pt_df=player_tracking_df.loc[player_tracking_df['position'].isin(rel_players)]
```

```python
pt_df.head()
```

```python
#length of data frame for relative players 
len(pt_df)
```

__The information below tells us that player tracking data is only given for defenders in coverage. Therefore, not all players on the field were tracked, so let's try and keep track of the number of people in coverage and who are defenders.__

```python
#For example, defensive line. 
pt_df[pt_df.position=='DL'].displayName.unique()
#This tells us that there is only one defensive lineman in coverage situation. 
```

```python
#This tells us that there were a few defensive ends in coverage situations but not all.  
pt_df[pt_df.position=='DE'].displayName.unique()
```

```python
#Make a new 'defender' column to dictate who is a defender
defenders = ['SS', 'FS', 'MLB', 'CB', 'LB', 'OLB', 'ILB', 'DL', 'DB', 'NT', 'S', 'DE', 'DT']
#pt_df['defender'] = pt_df.loc[pt_df['position'].isin(defenders)]
pt_df['defender'] = pt_df['position'].isin(defenders)
```

```python
#Count the number of defenders in coverage for a specific play. Make a new 'coverage_count' column. 
count_defenders = pt_df[(pt_df['defender'] == True) & (pt_df['frameId'] == 1)].groupby('new_Id').size().reset_index(name='coverage_count')
#count_defenders.defender_count.value_counts()
pt_df = pt_df.merge(count_defenders, on='new_Id')
```

```python
pt_df.head()
```

### Combine player tracking data and the plays data to have more features for predictions.

```python
#Recall what the plays data looks like: 
plays_df.head()
```

```python
#Combine the plays data and the player tracking data from week 1
new_df=pd.merge(pt_df, plays_df)
```
